```python
#computational columns
```

```python
inner_df4["RevenuePerKilloWatt"]=inner_df4["RevenueProduced"]/inner_df4["kilowatt_production"]
inner_df4["NetProfitForThe1stYearOfInstallation"]=inner_df4["RevenueProduced"]-inner_df4["yearlyupkeep"]
inner_df4["HoursPerYearOfOperation"]=inner_df4["HoursPerWeekOfOperation"]*52
inner_df4["MaintenanceCostPerYear"]=inner_df4["yearlyupkeep"]/inner_df4["HoursPerYearOfOperation"]
```

```python
conditions = [
    (inner_df4["MaintenanceCostPerYear"] > 1) & (inner_df4["kilowatt_production"] < 3),
    (inner_df4["MaintenanceCostPerYear"] > 0.5) & (inner_df4["MaintenanceCostPerYear"] < 1) & (inner_df4["kilowatt_production"] > 3) & (inner_df4["kilowatt_production"] < 6),
    (inner_df4["kilowatt_production"] > 6) & (inner_df4["kilowatt_production"] < 9)
]
choices= ["high","high","high"]  
inner_df4["Efficiency"]= np.select(conditions, choices , default='low')
```

```python

condition1 = (inner_df4['hasroads'] == 'Yes') & (inner_df4['haseasement'] == 'Yes') & (inner_df4['AreaOfExpertise'] == 'High Capacity')
condition2 = (inner_df4['hasroads'] == 'Yes') & (inner_df4['haseasement'] == 'Yes') & (inner_df4['AreaOfExpertise'] == 'High Efficiency')

good_conditions = condition1 | condition2

inner_df4['VendorQuality'] = 'Bad'
inner_df4.loc[good_conditions, 'VendorQuality'] = 'Good'
inner_df4[["hasroads","haseasement", "AreaOfExpertise", "VendorQuality"]]
```

```python
#aggregate_df
```

```python
aggregate_df= inner_df4.groupby('region_name').agg({
    "RevenuePerKilloWatt" : "mean",
    "HoursPerYearOfOperation" : "mean",
    "MaintenanceCostPerYear" : "mean"
})

aggregate_df.to_csv("region_aggregate.csv")
```

```python
#high efficiency model df
```

```python
HighEfficiencyDF= inner_df4[inner_df4["Efficiency"]=='high']
HighEfficiencyDF_Select_colums= HighEfficiencyDF[["model_cd","modelname","installcost","yearlyupkeep"]]
HighEfficiencyDF_Select_colums.to_csv("high_efficiency_model.csv", index=False)
```

```python
# Vendor preium column
```

```python
inner_df4['new_vendor_percentage'] = inner_df4['VendorPercent']

inner_df4.loc[inner_df4['VendorQuality'] == 'Good', 'new_vendor_percentage'] = inner_df4['VendorPercent'] * 1.05
inner_df4.loc[inner_df4['VendorQuality'] != 'Good', 'new_vendor_percentage'] = inner_df4['VendorPercent'] * 0.95

# Group by the specified columns and compute average
aggregate_df = inner_df4.groupby(['region_name', 'Vendor', 'VendorQuality']).agg({
    'VendorPercent': 'mean',
    'new_vendor_percentage': 'mean'
}).reset_index()

# Save the aggregated data to a CSV file
aggregate_df.to_csv("vendor_quality.csv", index=False)
```

```python
# Create updated_hour column based on region
inner_df4['updated_hour'] = inner_df4['HoursPerWeekOfOperation']
inner_df4.loc[inner_df4['State'] == 'TX', 'updated_hour'] = inner_df4['HoursPerWeekOfOperation'] * 1.06
inner_df4.loc[inner_df4['State'] == 'OK', 'updated_hour'] = inner_df4['HoursPerWeekOfOperation'] * 1.045
inner_df4.loc[inner_df4['State'] == 'NM', 'updated_hour'] = inner_df4['HoursPerWeekOfOperation'] * 1.04

# Calculate hour_diff
inner_df4['hour_diff'] = inner_df4['updated_hour'] - inner_df4['HoursPerWeekOfOperation']

# Filter data for high efficiency and compute the sum
aggregate_df = inner_df4[inner_df4['Efficiency'] == 'high'].groupby('modelname').agg({
    'HoursPerWeekOfOperation': 'sum',
    'updated_hour': 'sum',
    'hour_diff': 'sum'
}).reset_index()

# Save the aggregated data to a CSV file
aggregate_df.to_csv("operation_hours_raises.csv", index=False)
```
