```python
import io
import pandas as pd
import requests as r
```

```python
url = 'http://drd.ba.ttu.edu/isqs6339/ex/L2.1/boxdata.csv'
```

```python
# Fetch the CSV data from the URL
res = r.get(url)
```

```python
# Convert the fetched data into a pandas DataFrame
dfship = pd.read_csv(io.StringIO(res.text))
# Display the top 10 rows of the DataFrame
dfship.head(10)
```

```python
# Extract unique box-related data into dfbox and sort by boxtxt
dfbox = dfship[['boxtxt', 'width', 'length', 'depth', 'costtomake',
       'durability']].drop_duplicates().sort_values(by='boxtxt')
```

```python
# Update the box and store details in the main dfship DataFrame using data from dfbox, dfstore, and dfstoreend
for index, row in dfship.iterrows():
    #box updates
    dfship.at[index, 'width'] = dfbox.at[row['boxtxt'], 'width']
    dfship.at[index, 'length'] = dfbox.at[row['boxtxt'], 'length']
    dfship.at[index, 'depth'] = dfbox.at[row['boxtxt'], 'depth']
    dfship.at[index, 'costtomake'] = dfbox.at[row['boxtxt'], 'costtomake']
    dfship.at[index, 'durability'] = dfbox.at[row['boxtxt'], 'durability']
    
    #store updates - begin stores
    dfship.at[index, 'beginstoretxt'] = dfstore.at[row['beginstorenumber'], 'beginstoretxt']
    dfship.at[index, 'beginNightLoading'] = dfstore.at[row['beginstorenumber'], 'beginNightLoading']
    dfship.at[index, 'beginAutomation'] = dfstore.at[row['beginstorenumber'], 'beginAutomation']
    dfship.at[index, 'begininsurracecost'] = dfstore.at[row['beginstorenumber'], 'begininsurracecost']

    dfship.at[index, 'endstoretxt'] = dfstoreend.at[row['endstorenumber'], 'endstoretxt']
    dfship.at[index, 'endNightLoading'] = dfstoreend.at[row['endstorenumber'], 'endNightLoading']
    dfship.at[index, 'endAutomation'] = dfstoreend.at[row['endstorenumber'], 'endAutomation']
    dfship.at[index, 'endinsurracecost'] = dfstoreend.at[row['endstorenumber'], 'endinsurracecost']
    print(index, row['boxtxt'], row['beginstorenumber'], row['endstorenumber'])
```

```python
#Now, let's change the data types

dfship['width'] = pd.to_numeric(dfship['width'])
dfship['length'] = pd.to_numeric(dfship['length'])
dfship['depth'] = pd.to_numeric(dfship['depth'])
dfship['costtomake'] = pd.to_numeric(dfship['costtomake'])
dfship['durability'] = pd.to_numeric(dfship['durability'])
```

```python
dfship.loc[dfship['endNightLoading'] == 'No', 'e_nightloading'] = False
dfship.loc[dfship['endAutomation'] == 'No', 'e_automation'] = False
dfship.loc[dfship['beginNightLoading'] == 'No', 'b_nightloading'] = False
dfship.loc[dfship['beginAutomation'] == 'No', 'b_automation'] = False
```

```python
dfship['endNightLoading'] = dfship['e_nightloading']
dfship['endAutomation'] = dfship['e_automation']
dfship['beginNightLoading'] = dfship['b_nightloading']
dfship['beginAutomation'] = dfship['b_automation']
```

```python
#Now, let's answer the questions:
#What is the average weight of shipments originating from each store?
dfship[['beginstoretxt', 'weightofshipment']].groupby('beginstoretxt').mean()
```

```python
#How many stores allow for nightloading or automation?
#Notice, stores are not shared in this dataset for begin and end.  
#i.e. 2 separate piles of stores.
begstore = dfship[['beginstoretxt', 'beginNightLoading', 'beginAutomation']].drop_duplicates().sort_values(by='beginstoretxt')
endstore = dfship[['endstoretxt', 'endNightLoading', 'endAutomation']].drop_duplicates().sort_values(by='endstoretxt')

NL = len(begstore[begstore['beginNightLoading'] == True]) + len(endstore[endstore['endNightLoading'] == True])
Auto = len(begstore[begstore['beginAutomation'] == True]) + len(endstore[endstore['endAutomation'] == True])
print('Night Loading:  ', NL)
print('Automation:  ', Auto)
```
